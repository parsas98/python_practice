```python
def __equal_sides(number_of_sides : int) -> int :
    sides =[]
    ns =number_of_sides
    for num in range(1,ns+1) :
        sides.append(int(input(f"Please Enter {num}th Side of Your Shape :")))
    index_list = []
    n = 0
    for i in range(ns) :
        for j in range(ns) :
            if i != j :
                if sides[i]==sides[j] :
                    n += 1
                    b =[i,j]
                    index_list.append(b)
                else :
                    continue
            else :
                pass
    t = 0
    for i in range(n) :
        if index_list[0][0] == index_list[i][1] :
            t += 1
    number_of_equal_sides =int(n/t)
    return number_of_equal_sides
```
